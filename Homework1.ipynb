#Tokenization and Stop Word Removal - Step 3

Tokenize the text and remove stop words using NLTK. Also gather some stats such as total words, unique words, stop words, etc.

```python
# Text column
texts = df['text']
```

```python
# Define stop words
stop_words = set(stopwords.words('english'))
```

```python
# Initialize containers for words and stop words
all_words = []
all_unique_words = set()
all_stop_words = []
unique_stop_words = set()
```

```python
# Tokenize and process the text
for text in texts:
    words = nltk.word_tokenize(text)
    all_words.extend(words)
    all_unique_words.update(words)

    # Filter
    stop_words_in_text = [word for word in words if word.lower() in stop_words]
    all_stop_words.extend(stop_words_in_text)
    unique_stop_words.update(stop_words_in_text)

# Display results for verification
{
    'Total number - words': len(all_words),
    'Total number - unique words': len(all_unique_words),
    'Total number - stop words': len(all_stop_words),
    'Total number - unique stop words': len(unique_stop_words)
}
```

#Frequency Calculation - Step 4

Calculate the word frequencies using counter from the collections library.

```python
# Calculate word frequencies
word_freq = Counter(all_words)
word_freq_without_stop_words = Counter(word for word in all_words if word.lower() not in stop_words)
```

```python
# Results (display the 10 most common words)
results = {
    'Frequency - each unique word': word_freq.most_common(10),
    'Frequency - each unique word without stop words': word_freq_without_stop_words.most_common(10)
}
```

```python
# Print results
for key, value in results.items():
    print(f'{key}: {value}')
```

We use counter to find the most common words and their counts, both with and without stop words
